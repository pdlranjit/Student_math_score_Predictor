```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import  train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python

data=pd.read_csv('StudentsPerformance.csv')
```

```python
numeric_features=['reading score','writing score']

categorical_features=['gender','race/ethnicity','parental level of education','lunch','test preparation course']
```

```python
numeric_features
```

```python
X=data[numeric_features + categorical_features]
Y=data['math score']
```

```python
X
```

```python
X.head()
Y.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numeric_features),  # Keep numeric as-is
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_features)  # Encode categorical
    ])
```

```python
# This is like WRITING the recipe/formula
model = Pipeline([
    ('preprocessor', preprocessor),     # "Step 1: Prepare ingredients"
    ('regressor', LinearRegression())   # "Step 2: Apply this cooking method"
])

# It's just instructions, no actual cooking yet!
```

```python
# This is like USING the recipe with REAL ingredients
model.fit(X_train, Y_train)
# "Take THESE specific ingredients (X_train) 
#  and learn how they make THIS specific dish (y_train)"
```

```python
X_train_transformed = preprocessor.transform(X_train)
print("Shape after ColumnTransformer:", X_train_transformed.shape[1])
```

```python
preprocessor
```

```python
numeric_features
categorical_features
```

```python
predictions=model.predict(X_test)
```

```python
predictions
```

```python
preprocessor.fit(X)

# Get the one-hot encoded categorical column names
cat_columns = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)

# Combine with numeric columns
all_columns = numeric_features + list(cat_columns)

print("Columns after ColumnTransformer / Pipeline:", all_columns)
print("Total number of features:", len(all_columns))
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
score=r2_score(Y_test,predictions)
```
